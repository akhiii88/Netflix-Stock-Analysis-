# 📈 Netflix Stock Price Analysis
Analyze how Netflix stock behaved during real-world events like the COVID-19 pandemic.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../Data/nflx_stock_data.csv', parse_dates=['Date'])
df.set_index('Date', inplace=True)
df.head()
```

```python
# Plot closing price
plt.figure(figsize=(12, 5))
plt.plot(df['Close'], label='Close Price')
plt.axvline(pd.to_datetime('2020-03-01'), color='red', linestyle='--', label='COVID Start')
plt.title('Netflix Stock Price Over Time')
plt.legend()
plt.show()
```
